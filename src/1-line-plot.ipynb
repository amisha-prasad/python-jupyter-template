### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0, 11)          
y = x ** 2                    


plt.plot(x, y, marker='o')


plt.title("Line Plot of x vs x²")  
plt.xlabel("x values")              
plt.ylabel("y = x²")               
plt.grid(True)                      

# Show the plot
plt.show()
```
